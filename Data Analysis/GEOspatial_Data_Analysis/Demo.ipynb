# Using Contextily import to overlay maps: 
### EPSG:
code is a unique identifier for different coordinate systems. It's a public registry of geodetic datums, spatial reference systems, Earth ellipsoids, coordinate transformations and related units of measurement.

- Defining web marcator to overlay data onto a actual map 

```python
data_web_marcator= Race.to_crs(epsg = 3857)
data_web_marcator

fig, ax_layered = plt.subplots(figsize = (15,15))

data_web_marcator[data_web_marcator['Percent Not Hispanic or Latinos']>90].plot(ax=ax_layered, alpha = 0.8, cmap = 'RdYlGn_r')

ax_layered.axis("off")

ax_layered.set_title('90% of the Popultion Not Hispanic or Latino', fontsize=16)

ctx.add_basemap(ax_layered)
```
